## Step 1. Import Libraries and Data 

```python
import pandas as pd 
import numpy as np

data = pd.read_csv('autos.csv')
data
```

### 2. Feature Extraction 


```python
# Extracting Date and Time Features

# Convert 'dateCrawled' to datetime format
df['dateCrawled'] = pd.to_datetime(df['dateCrawled'])

# Extract new features from 'dateCrawled'
df['hour'] = df['dateCrawled'].dt.hour
df['day_of_week'] = df['dateCrawled'].dt.dayofweek
df['month'] = df['dateCrawled'].dt.month

# Display the first few rows with the new features
print(df[['dateCrawled', 'hour', 'day_of_week', 'month']].head())
```

```python
df['seller'].unique()
```

```python
# Apply One-Hot Encoding on the seller column
df = pd.get_dummies(df, columns = ['seller'], prefix = 'seller')

# Display the df with new features 
print(df.head())
```

```python
df1 = df[['price', 'yearOfRegistration', 'kilometer', 'postalCode']]
df1.head()

```

## 3. Feature Selection 

```python
# Removing Redundant Features
import seaborn as sns
import matplotlib.pyplot as plt


# Calculate the correlation matrix 
corr_matrix = df1.corr()

# Plot the corelation heatmap
plt.figure(figsize = (10,8))
sns.heatmap(corr_matrix, annot = True, cmap = 'coolwarm')
plt.show()
```

```python
# Recusive Feature Elimination (RFE)
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import RFE

# Define the feature matrix(x) and target variable(y)
x = df1.drop(columns = ['price'])
y = df1['price']
```

```python
# Initialize a linear regression model 
model = LinearRegression()

# Apply Recursive Feature Elimination(RFE)
rfe = RFE(model, n_features_to_select = 2)
fit = rfe.fit(x,y)

selected_features = x.columns[fit.support_]
print("Selected Features: ", selected_features)
```

## 4. Model Building 

```python
from sklearn.model_selection import train_test_split

# Split the data into train(80%) and test(20%) set
x_train, x_test, y_train, y_test = train_test_split(x[selected_features], y, test_size = 0.20, random_state = 42)
```

```python
# Train the Model 
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error

# Initialize the Random Forest regressor
model = RandomForestRegressor(n_estimators = 100, random_state = 42)

# Train the model
model.fit(x_train, y_train)

# Predict on the test set
y_pred = model.predict(x_test)

# Evaluate the Model
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error : {mse}")
```
