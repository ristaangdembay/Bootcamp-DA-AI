```python
import pandas as pd
import numpy as np

#Load the datasets
data = {
    'ID': [1, 2, 3, 4, 5, 6, 7, 7],
    'Name': ['Alice', 'Bob', 'Charlie', np.nan, 'Eve', 'Frank', 'Grace', 'Grace'],
    'Age': [25, np.nan, 30, 45, 'unknown', 35, 30, 30],
    'Salary': [50000, 60000, np.nan, 70000, 80000, 'not available', 70000, 70000],
    'Department': ['HR', 'Finance', 'HR', 'IT', np.nan, 'Finance', 'IT', 'IT']
}

df = pd.DataFrame(data)
df.head(8)
```

```python
missing_values = df.isnull().sum()
missing_values
```

```python
df['Name'].fillna('Unkown', inplace = True) #fillna() --> find NaN and replace eith another value

# Display the dataframe after replacing NaN with 'unkown'
df.head()
```

```python
#df['head'].fillna('Unknown', inplace = True)

df['Age'] = pd.to_numeric(df['Age'], errors = 'coerce')   #coverts to numeric, else fill with NaN
df
```

```python
df['Salary'] = pd.to_numeric(df['Salary'], errors = 'coerce')
df
```

```python
df.dtypes
```

```python
df = df.drop_duplicates(subset = 'ID')

df.head()
```
