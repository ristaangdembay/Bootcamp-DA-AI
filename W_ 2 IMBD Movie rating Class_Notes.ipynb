```python
movie_container = soup.find_all('div', class_ = "movie-container")
```

```python
#Extract movie details

movies = []
for container in movie_container:
    movie_title = container.find('h1', class_ = 'movie-title').text.strip()
    release_date = container.find("p").text.strip().replace('Released on:', '').strip()
    imdb_rating = container.find("p", class_ = 'ratings')#strip().replace('IMDb Rating:', '')#.strip()#.split('/')[0]
    
    #imdb_rating = container.find("span", text = 'IMDb Rating:').find_next_sibling(text = True).split('/')#[0]
    
    
    movies.append({
        "Movies Title": movie_title,
        "Release Date": release_date,
        "IMDb Raing": imdb_rating
    })
    
```

```python
movies

```

```python
#Convert a DataFrame

df = pd.DataFrame(movies)
df
```

```python
# Saving the data to a csv file 
df.to_csv("movies.csv", index =  False)
```
